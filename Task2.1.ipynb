```python
#Ipotrt libraraies
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
import scipy.stats as ss
from itertools import combinations
```

```python
filepath = r'C:\Users\Administrator\Documents\Grad School\Statistics & Machine Learning\Discussion Task Week2\Fifa_Dataset.csv'
data_df = pd.read_csv(filepath)
#Remove column 24 and 25, Not intersted in cleaning those and using it for our analysis
data_df = data_df[['preferred_foot','nationality','age','club_team','value_euro','wage_euro','release_clause_euro', 'height_cm','weight_kgs','crossing', 'finishing', 'heading_accuracy', 'short_passing', 'volleys', 'dribbling', 'curve', 'freekick_accuracy', 'long_passing', 'ball_control', 'acceleration', 'sprint_speed', 'agility', 'reactions', 'balance', 'shot_power', 'jumping', 'stamina', 'strength', 'long_shots', 'aggression', 'interceptions', 'positioning', 'vision', 'penalties', 'composure', 'marking', 'standing_tackle', 'sliding_tackle', 'GK_diving', 'GK_handling', 'GK_kicking', 'GK_positioning', 'GK_reflexes']]
data_df = pd.DataFrame(data_df)
#replace empty column values with average
#replace our values with zeroes with averages
data_df['value_euro'] = data_df['value_euro'].fillna(data_df['value_euro'].mean())
data_df['wage_euro'] = data_df['wage_euro'].fillna(data_df['wage_euro'].mean())
data_df['release_clause_euro'] = data_df['release_clause_euro'].fillna(data_df['release_clause_euro'].mean())
data_df.info()
```

```python
#seperate our columns
preferred_Right_df = data_df[data_df.preferred_foot == 'Right']

preferred_Left_df = data_df[data_df.preferred_foot == 'Left']
#create just numberical data
preferred_Right_num_df = preferred_Right_df.drop(columns=['preferred_foot', 'club_team','nationality'])
preferred_Left_num_df = preferred_Left_df.drop(columns=['preferred_foot', 'club_team','nationality'])
```

```python
#re create a Dataframe
averages_foot = pd.DataFrame(
    {'atribute': colum_name,
     'average_right': Average_right,
     'average_left': Average_left
    })
averages_foot
```

```python
#Create the first column of our base data frame
country_combo_df = pd.DataFrame(colum_name) # this is just a dataframe with a list of our atributes
```

```python
for country in countrylist: #there is probably an easier way to do this. I know a pivot table can get the average for each country on excel with ease
    tempmeanlst = []
    temp_fifa1 = data_df[data_df.nationality == country]
    temp_fifa1 = temp_fifa1.drop(columns=['club_team','preferred_foot','nationality'])
    for column in temp_fifa1:
        average = temp_fifa1[column].mean()
        tempmeanlst.append(round(average, 2))
    country_string = country + "_mean"
    country_combo_df[country_string] = tempmeanlst
country_combo_df = country_combo_df.rename(columns={0:"Atribute"})
```

```python
for combo in combinations(countrylist, 2):
    templist_statistic = [] #reset our lists that we're adding to
    templist_pvalue = []
    templist2_df = []
    temp_fifa1 = data_df[data_df.nationality == combo[0]]
    temp_fifa1 = temp_fifa1.drop(columns=['club_team','preferred_foot','nationality'])
    temp_fifa2 = data_df[data_df.nationality == combo[1]]
    temp_fifa2 = temp_fifa2.drop(columns=['club_team','preferred_foot','nationality'])
    for column in temp_fifa1:
        result = ttest_ind(temp_fifa1[column], temp_fifa2[column])
        templist_statistic.append(result.statistic)
        templist_pvalue.append(result.pvalue)
        templist2_df.append(result.df)
    statistic_col_name = combo[0] + "_vs._" + combo[1] + "_statistic"
    pvalue_col_name = combo[0] + "_vs._" + combo[1] + "_Pvalue"
    df_col_name = combo[0] + "_vs._" + combo[1] + "_df"
    country_combo_df[statistic_col_name] = templist_statistic
    country_combo_df[pvalue_col_name] = templist_pvalue
    country_combo_df[df_col_name] = templist2_df
```
